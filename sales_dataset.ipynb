```python
import numpy as np

# Creating Sales Dataset
# Columns: Quantity Sold, Unit Price

quantity_sold = np.array([10, 5, 8, 12, 7])
unit_price = np.array([50.0, 100.0, 75.0, 60.0, 90.0])

# Calculate Total Sales for each product
total_sales = quantity_sold * unit_price

# Display Dataset
print("Quantity Sold:", quantity_sold)
print("Unit Price:", unit_price)
print("Total Sales:", total_sales)

# Calculate Overall Sales
overall_sales = np.sum(total_sales)
print("Overall Sales:", overall_sales)
```
